```python
#지피티 힌트 -> 백트래킹 + DFS

def search(n, total):
    global answer

    # 최솟값보다 커지면 더 볼 필요 없음
    if answer <= total:
        return
    
    # 끝까지 다 선택했을 때
    if n == N:
        answer = min(answer, total)
        return

    # 아직 안 고른 열 중 하나 선택
    for j in range(N):
        if visited[j] == 0:
            visited[j] = 1
            search(n + 1, total + arr[n][j])
            visited[j] = 0

T = int(input())
for test_case in range(T):
    N = int(input())
    arr = [list(map(int, input().split())) for _ in range(N)]
    answer = 100 #최대로 커질 수 있는
    visited = [0] * N
    search(0, 0)

    print(f"#{test_case + 1} {answer}")
```
