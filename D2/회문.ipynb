### [파이썬 S/W 문제해결 기본] 3일차 - 회문

ABBA처럼 어느 방향에서 읽어도 같은 문자열을 회문이라 한다. NxN 크기의 글자판에서 길이가 M인 회문을 찾아 출력하는 프로그램을 만드시오.

회문은 1개가 존재하는데, 가로 뿐만 아니라 세로로 찾아질 수도 있다.
 

예를 들어 N=10, M=10 일 때, 다음과 같이 회문을 찾을 수 있다.
 


```python
import copy

#회문 찾기 함수
def find_answer(arr_list):
    ans = 0

    for j in range(N):
        sample = arr_list[j]

        for i in range(N-M+1):
            not_reversed = ''.join(sample[i:i+M])
            #print(not_reversed)
            reversed = ''.join(sample[i:i+M][::-1])
            #print(reversed)
            if not_reversed == reversed:
                ans = not_reversed
                break
            else:
                continue
        if ans == 0:
            continue
        else:
            break
    
    return ans

#시작
T = int(input())

for _ in range(T):
    first_list = []
    N, M = map(int, input().split())

    for i in range(N):
        first_list.append(list(input()))
    
    # 가로 회문 찾기
    ans = find_answer(first_list)
            
    # 가로에서 못 찾았을 때 세로 회문 찾기
    if ans == 0:
        #전치행렬 만들기
        second_list = copy.deepcopy(first_list)
        
        for i in range(N):
            for j in range(N):
                second_list[i][j] = first_list[j][i]

        ans = find_answer(second_list)
    else:
        pass
    
    print(f'#{_+1}', ans)

```
