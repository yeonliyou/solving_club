### 자료 구조 - stack



```python
#eval() 안쓰는 버전
#예외처리 : 0으로 나누는 경우, 연산자가 없는 경우, 연산할 숫자가 부족한 경우

T = int(input())
oper = ['/', '//', '+', '*','-', '**', '%']

#연산하는 함수 만들기
def calculate_str(one,oper,two):
    one = int(one)
    two = int(two)
    if oper == '/':
        value = one/two
    elif oper == '//':
        value = one//two
    elif oper == '+':
        value = one+two
    elif oper == '*':
        value = one*two
    elif oper == '-':
        value = one-two
    elif oper == '**':
        value = one**two
    elif oper == '%':
        value = one%two

    return value

for i in range(T):
    input_list = list(input().split())
    #print(input_list)

    a = False
    while a == False:
        for idx, j in enumerate(input_list):
            #연산자면 계산
            if j in oper:
                try:
                    one = input_list[idx-2]
                    two = input_list[idx-1]
                    cal = calculate_str(one,j,two)
                    #print(cal)

                    #연산에 사용한 연산자와 숫자들은 삭제
                    input_list.pop(idx)
                    input_list.pop(idx-1)
                    input_list.pop(idx-2)

                    #새롭게 연산된 수는 input_list에 추가
                    input_list.insert(idx-2, str(cal))
                    #print(input_list)
                    break
                #one, two가 둘다 제대로 존재x 등의 => 연산을 할 수 없을 때
                except:
                    result = 'error'
                    a = True
                    break
            # '.' 차례가 오면 반복문 종료
            elif j == '.':
                a = True
                result = input_list[0]
                #break
            #숫자면 그냥 통과
            else:
                try:
                    if type(int(j)) == int:
                        continue
                except:
                    result = 'error'
                    a = True
                    break

        
    print(f'#{i+1}',result)
```
