```python
#eval()  
# : 0  ,   ,    

T = int(input())
oper = ['/', '//', '+', '*','-', '**', '%']

#  
def calculate_str(one,oper,two):
    one = int(one)
    two = int(two)
    if oper == '/':
        value = one/two
    elif oper == '//':
        value = one//two
    elif oper == '+':
        value = one+two
    elif oper == '*':
        value = one*two
    elif oper == '-':
        value = one-two
    elif oper == '**':
        value = one**two
    elif oper == '%':
        value = one%two

    return value

for i in range(T):
    input_list = list(input().split())
    #print(input_list)

    a = False
    while a == False:
        for idx, j in enumerate(input_list):
            # 
            if j in oper:
                try:
                    one = input_list[idx-2]
                    two = input_list[idx-1]
                    cal = calculate_str(one,j,two)
                    #print(cal)

                    #    
                    input_list.pop(idx)
                    input_list.pop(idx-1)
                    input_list.pop(idx-2)

                    #   input_list 
                    input_list.insert(idx-2, str(cal))
                    #print(input_list)
                    break
                #one, two   x  =>     
                except:
                    result = 'error'
                    a = True
                    break
            # '.'    
            elif j == '.':
                a = True
                result = input_list[0]
                #break
            #  
            else:
                try:
                    if type(int(j)) == int:
                        continue
                except:
                    result = 'error'
                    a = True
                    break

        
    print(f'#{i+1}',result)
```
