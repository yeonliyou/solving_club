```python
#t   
T = int(input())

for i in range(T):
    V, E = map(int, input().split())
    subgraph_list = []
    final_list = []

    start_node = []
    end_node = []
    #   
    for _ in range(E):
        #subgraph_list.append()
        x = list(input().split())
        subgraph_list.append(x)
        start_node.append(x[0])
        end_node.append(x[1])

    S, G = input().split()

    #final 
    for idx in range(E):
        if start_node[idx] in end_node:
            sam_idx = end_node.index(start_node[idx])
            final_list.append([subgraph_list[sam_idx], start_node[idx]])
        else:
            final_list.append(subgraph_list[idx])

    anw = 0

    for j in final_list:
        if j[0] == S and j[-1] == G:
            anw = 1
            break
        else:
            continue

    print(f'#{i+1}', anw)
```
