### [파이썬 S/W 문제해결 기본] 4일차 - 그래프 경로

V개 이내의 노드를 E개의 간선으로 연결한 방향성 그래프에 대한 정보가 주어질 때, 특정한 두 개의 노드에 경로가 존재하는지 확인하는 프로그램을 만드시오.

두 개의 노드에 대해 경로가 있으면 1, 없으면 0을 출력한다.
 

예를 들어 다음과 같은 그래프에서 1에서 6으로 가는 경로를 찾는 경우, 경로가 존재하므로 1을 출력한다.



```python
#t실시간으로 점차 만드는 것
T = int(input())

for i in range(T):
    V, E = map(int, input().split())
    subgraph_list = []
    final_list = []

    start_node = []
    end_node = []
    #그래프 연결 정보 담기
    for _ in range(E):
        #subgraph_list.append()
        x = list(input().split())
        subgraph_list.append(x)
        start_node.append(x[0])
        end_node.append(x[1])

    S, G = input().split()

    #final 만들기
    for idx in range(E):
        if start_node[idx] in end_node:
            sam_idx = end_node.index(start_node[idx])
            final_list.append([subgraph_list[sam_idx], start_node[idx]])
        else:
            final_list.append(subgraph_list[idx])

    anw = 0

    for j in final_list:
        if j[0] == S and j[-1] == G:
            anw = 1
            break
        else:
            continue

    print(f'#{i+1}', anw)
```
