###  : 
-  (Binary Search)        .           ,      .    (Divide and Conquer)      .
-  
  1.  :     ** **  .     ,       .
  2.  :    **   **.  0   ,   (left + right) / 2   .
  3.   :     (target) .  ,      .
       -       :  .
       -       :           .
       -       :           .
  4.  :     2 3             .

```python
import copy

def calculate_count(P, m):
    l = 1
    r = P
    count_value = 0
    c = 0 
    while c != m:
        c = int((l+r)/2)
        #  <  
        if c < m:
            l = c
            r = r
        #  >   
        else:
            r = c
            l = l
        count_value += 1
    return count_value

T = int(input())

for i in range(T):
    P, A, B = map(int, input().split())
    #A 
    a_count = calculate_count(P, A)
    #B 
    b_count = calculate_count(P, B)

    if a_count > b_count:
        print(f'#{i+1}', 'B')
    elif a_count < b_count:
        print(f'#{i+1}', 'A')
    else:
        print(f'#{i+1}', 0)
```
