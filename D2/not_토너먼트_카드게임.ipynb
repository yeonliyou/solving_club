### [파이썬 S/W 문제해결 기본] 5일차 - 토너먼트 카드게임




```python
T = int(input())


def who_win(students, N):
    #첫 index_list 초기화 -> 0, 1, ...., N-1
    index_list = [i for i in range(N)]
    print(index_list)

    while N > 5:
        index_list_1 = []
        index_list_2 = []
        group1 = students[:(N-1)//2+1]
        group2 = students[(N-1)//2+1:]

        index_list_1 = index_list[:(N-1)//2+1]
        index_list_2 = index_list[(N-1)//2+1:]

        #이긴 사람 가위바위보 리스트
        group1_win = []
        group2_win = []

        filter_index_list_1 = []
        for v, idx in zip(group1, index_list_1):
            if v == min(group1):
                group1_win.append(v)
                index_list_1.append(idx)
            else:
                continue
        
        filter_index_list_2 = []
        for m, idx in zip(group2, index_list_2):
            if m == min(group2):
                group2_win.append(m)
                index_list_2.append(idx)
            else:
                continue
        
        index_list_1 = filter_index_list_1
        index_list_2 = filter_index_list_2
        students = group1_win.extend(group2_win)
        index_list = filter_index_list_1.extend(filter_index_list_2)

        students = group1_win.extend(group2_win)
        index_list = index_list_1.extend(index_list_2)
        print(students)
        print(index_list)
        N = len(students)
    
    while N > 2:
        #group1 체크
        group1 = students[:(N-1)//2+1]
        group2 = students[(N-1)//2+1:]
        print('그룹1, 그룹2', group1, group2)

        index_list_1 = []
        index_list_2 = []

        #인덱스 모음 리스트 만들기
        index_list_1 = index_list[:(N-1)//2+1] #마지막 인덱싱은 포함 안됨
        index_list_2 = index_list[(N-1)//2+1:]
        print('인덱스 모음', index_list_1, index_list_2)

        #이긴 사람 가위바위보 리스트
        group1_win = []
        group2_win = []

        filter_index_list_1 = []
        for v, idx in zip(group1, index_list_1):
            if v == min(group1):
                group1_win.append(v)
                filter_index_list_1.append(idx)
            else:
                continue
        
        print(group1_win, index_list_1)

        if len(group1_win) == 1:
            pass
        else:
            group1_win = [group1_win[0]]
            filter_index_list_1 = [filter_index_list_1[0]]

        filter_index_list_2 = []
        #그룹2 체크
        for m, idx in zip(group2, index_list_2):
            if m == min(group2):
                group2_win.append(m)
                filter_index_list_2.append(idx)
            else:
                continue
        
        print(group2_win, index_list_2)

        if len(group2_win) == 1:
            pass
        else:
            group2_win = [group2_win[0]]
            filter_index_list_2 = [filter_index_list_2[0]]
        
        index_list_1 = filter_index_list_1
        index_list_2 = filter_index_list_2
        students = group1_win.extend(group2_win)
        index_list = filter_index_list_1.extend(filter_index_list_2)
        print(students)
        print(index_list)
        N = len(students)

    return students[0]


for _ in range(T):
    N = int(input())
    students = []
    students.extend(list(map(int, input().split())))

    print(who_win(students, N))


    
```
