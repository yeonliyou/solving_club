### [파이썬 S/W 문제해결 기본] 4일차 - 종이붙이기



```python
T = int(input())

for _ in range(T):
    N = int(input())
    a = N//10
    #print(a)
    #N의 앞자리가 홀수 일 때
    if a%2 != 0:
        seq = [1]
        a1 = 1
        for i in range(a//2):
            a1 = a1*4 + 1
            seq.append(a1)
        
        anw = seq[-1]
    #N의 앞자리가 짝수 일 때
    else:
        seq = [3]
        b1 = 3
        for i in range(a//2 -1):
            b1 = b1*4 - 1
            seq.append(b1)

        anw = seq[-1]
    
    #print(seq)
    print(f'#{_+1}', anw)
```
